```python
from pandas_datareader import data

df = data.get_data_yahoo("AAPL")
df['return'] = df['Adj Close'].pct_change().fillna(0)
```

```python
import pandas_montecarlo
mc = df['return'].montecarlo(sims=10, bust=-0.1, goal=1)
```

```python
mc.plot(title="AAPL Returns Monte Carlo Simulations")  # optional: , figsize=(x, y)
```

```python
print(mc.stats)
```

```python
print(mc.maxdd)
```
